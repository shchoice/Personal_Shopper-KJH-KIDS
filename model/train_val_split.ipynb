```python
import os
import shutil

directory='masked_moved2'
folder_list=os.listdir(directory)
```

```python
print(len(folder_list))
```

```python
cnt=1
for folder in folder_list:    
    file_list=os.listdir(directory+'/'+folder+'/'+folder)
    length=len(file_list)
    
    if length>9:
        train=file_list[:int(length*9/10)]
        val=file_list[int(length*9/10):]
        train_path='train/'+folder+'/'+folder+'/'
        val_path='val/'+folder+'/'+folder+'/'
        if not os.path.exists(train_path):
            os.makedirs(train_path)
        if not os.path.exists(val_path):
            os.makedirs(val_path)            
        for t in train:
            shutil.copy(directory+'/'+folder+'/'+folder+'/'+t, train_path+t)
            
        for v in val:
            shutil.copy(directory+'/'+folder+'/'+folder+'/'+v, val_path+v)
        
        if cnt%100==0:
            print(str(cnt)+'/'+str(len(folder_list)))
        cnt+=1
```

```python
directory='val'
folder_list=os.listdir(directory)
```

```python
cnt=1
for i in folder_list:
    train_list=os.listdir('train/'+i+'/'+i)
    val_list=os.listdir(directory+'/'+i+'/'+i)
    print(cnt,i, len(train_list),'/',len(train_list)+len(val_list))
    print(cnt,i, len(train_list)/(len(train_list)+len(val_list))*100,'%')
    cnt+=1
```
